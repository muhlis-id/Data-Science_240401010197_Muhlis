LIST OPERATION

```python
# Membuat List
data = [10, 20, 'Python', True, 3.14]

# Akses & Slicing
print(data[0])          # > 10
print(data[-1])         # > 3.14 (dari belakang)
print(data[1:3])        # > [20, 'Python']

#Manipulasi
data.append(99)         # Tambah di akhir
data.insert(1, 55)      # Sisipkan di posisi 1
data.remove(20)         # hapus nilai pertama = 20
#data.sort()             # urutkan (jika type sama)
data.sort(key=str)      # alternatif menggunakan string
```

```python
# DICTIONARY OPERATION

mahasiswa = {
    'nama': 'Budi Santoso',
    'nim': '2024001',
    'prodi': 'Informatika',
    'ipk': '3.85'
}

#Akses Nilai
print(mahasiswa['nama'])
print(mahasiswa.get('usia', 0))     # default jika key tidak ada

#Update dan hapus
mahasiswa['angkatan'] = 2024        # tambah key angjatan
del mahasiswa['nim']                # hapus key
```
